```python
#based on the population size organize the world countries into three size group.
df_y.loc[df_y['population'] > 40000000,'country_size'] = 'L'
df_y.loc[(df_y['population']> 1000000)&(df_y['population']<40000000),'country_size']='M'  
df_y.loc[df_y['population'] < 10000000,'country_size'] = 'S'
```

```python
fig,ax=plt.subplots(figsize=(7,7))
df_y_1960=df_y[df_y['year']==1960]
ax= sns.scatterplot(data=df_y_1960, y='fertility_rate',x='life_expec',hue='country_size',
                    size='country_size',size_order=['L','M','S'],sizes=(100,300))
plt.title("Fertility_rate and life_expection in year 1960-- country size view",fontsize=12)
```
