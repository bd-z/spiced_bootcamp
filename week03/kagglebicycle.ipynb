```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns

df =pd.read_csv('C:/Users/zhang/spiced_bootcamp/unsupervised-lemon-encounter-notes/week_03/data/train.csv', parse_dates=True,index_col=[0])
df=df.drop(['casual','registered'],axis=1)
df.head()


def FindX(df,y):
    X=[]
    for i in df.columns:
        X.append(i)
    X.remove(y)
    print (X)
    return df[X]
y='count'

FindX(df,y)
```

```python
X_test=pd.DataFrame(X_test)
X_test


def DateTime(df,cl_name):
    df=df.reset_index()
    df[cl_name]=pd.to_datetime(df[cl_name])
    wek=[]
    hur=[]
    mth=[]
    for i in range(len(df)):
        wek.append(df[cl_name][i].weekday())
        hur.append(df[cl_name][i].hour)
        mth.append(df[cl_name][i].month)
    df['weekday']=wek
    df['hour']=hur
    df['month']=mth
    return df

X_train_d=DateTime(X_train,'datetime')
X_test_d=DateTime(X_test,'datetime')

X_kg_d=DateTime(X_kg,'datetime')

```

```python

def binning(df):
    binl=['early','morning','day','eve','night']
    binv=[-1,6.5,8.5,16.0,20.0,25.0]
    df.loc[:,'hour_bin'] = pd.cut(df['hour'],bins=binv,labels=binl,right=False)
    return df

```

```python
X_train_db=binning(X_train_d)
X_test_db=binning(X_test_d)
X_kg_db=binning(X_kg_d)
```
